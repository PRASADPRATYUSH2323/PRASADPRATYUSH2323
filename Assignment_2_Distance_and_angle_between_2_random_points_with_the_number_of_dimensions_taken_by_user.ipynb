# Section 1 taking numbers between 0 and 1

```python
# importing the libraries
import numpy as np
from numpy import random
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.ticker as ticker
```

```python
# typing the number of inputs
n=int(input('type the number of entries :'))
```

```python
# taking random data with the number of dimensions as the input take
a=np.random.rand(n,1)
b=np.random.rand(n,1)
print(a)
print()
print(b)
```

```python
# distance of vector a and x from origin
ssa=0
ssb=0
ssab=0
for i in range(0,n):
    ssa+=((a[i])**2)
    ssb+=((b[i])**2)
    ssab+=((a[i])-(b[i]))**2
dist_ao=ssa**.5
dist_bo=ssb**.5
dist_ab=ssab**.5
print('distance of a from origin',dist_ao)
print('distance of b from origin',dist_bo)
print('distance of b from a',dist_ab)
```

```python
# finding dot product of both points
A=np.array(a)
B=np.array(b)
C=A*B
print(C)
```

```python
# finding the sum of dot products 
sc=0
for i in C:
    sc+=i
print(sc)    
```

```python
# printing the cosine of angle
cost=sc/(dist_ao*dist_bo)
print(cost)
```

```python
# printing angle in degree and radians
print('angle in radians:',np.arccos(cost))
print('angle in degrees:',(np.arccos(cost))*180/np.pi)
```

# Section 2 taking numbers between -1 and 1

```python
# importing the libraries
import numpy as np
from numpy import random
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.ticker as ticker
```

```python
# typing the number of inputs
n=int(input('type the number of entries :'))
```

```python
# taking random data with the number of dimensions as the input take
a=np.random.rand(n,1)
b=np.random.rand(n,1)
a=(2*a)-1
b=(2*b)-1
print(a)
print()
print(b)
```

```python
# distance of vector a and x from origin 
ssa=0
ssb=0
ssab=0
for i in range(0,n):
    ssa+=((a[i])**2)
    ssb+=((b[i])**2)
    ssab+=((a[i])-(b[i]))**2
dist_ao=ssa**.5
dist_bo=ssb**.5
dist_ab=ssab**.5
print('distance of a from origin',dist_ao)
print('distance of b from origin',dist_bo)
print('distance of b from a',dist_ab)
```

```python
# finding dot product of both points
A=np.array(a)
B=np.array(b)
C=A*B
print(C)
```

```python
# finding the sum of dot products 
sc=0
for i in C:
    sc+=i
print(sc)    
```

```python
# printing the cosine of angle
cost=sc/(dist_ao*dist_bo)
print(cost)
```

```python
# printing angle in degree and radians
print('angle in radians:',np.arccos(cost))
print('angle in degrees:',(np.arccos(cost))*180/np.pi)
```
